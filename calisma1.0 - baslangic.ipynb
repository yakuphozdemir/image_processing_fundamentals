```python
path10 = r"C:\Users\yakup\Desktop\image_proccess_exercises\example photos\photo.jpg"
path11 = r"C:\Users\yakup\Desktop\image_proccess_exercises\example photos\photo2.jpg"
pic10 = cv.imread(path10)
pic11 = cv.imread(path11)

cv.imshow("photo",pic10)
cv.imshow("photo2",pic11)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
"""
#other way
a = int((pic10.shape[0] - pic11.shape[0])/2)
pic11 = cv.copyMakeBorder(pic11,a,a+1,0,0,borderType= cv.BORDER_REPLICATE)
print(pic11.shape)

b = int((pic11.shape[1]-pic10.shape[1])/2)
pic10 = cv.copyMakeBorder(pic10,0,0,b,b+1,cv.BORDER_REPLICATE)
print(pic10.shape)
cv.imshow("duz_porkatal",pic10)
cv.imshow("yan_porkatal",pic11)
cv.waitKey(0)
"""
```

Görüntü Piramitleri

* Bir Resmin Genişlik ve Yüksekliğini İki Katına Çıkarmak ya da Yarıya Düşürmek

```python
pic15 = pic1.copy()
pic16 = cv.pyrUp(pic15)
pic17 = cv.pyrDown(pic15)

print("orijinal :                       ",pic15.shape)
print("buudları iki katına çıkarılan:   ",pic16.shape)
print("buudları yarıya indirilen:       ",pic17.shape)

cv.imshow("pic15",pic15)
cv.imshow("pic16",pic16)
cv.imshow("pic17",pic17)
cv.waitKey(0)
cv.destroyAllWindows()
```
